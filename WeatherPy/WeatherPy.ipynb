```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from scipy.stats import linregress
from matplotlib import cm

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# View format of information for accessing data
url1 = f"http://api.openweathermap.org/data/2.5/weather?q={cities[0]}&units=imperial&appid={weather_api_key}"

response = requests.get(url1).json()   

print(json.dumps(response, indent=4, sort_keys=False))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create variables to store the necessary data
lat = []
lon = []
maxTemp = []
humidity = []
cloudiness = []
windSpeed = []
country = []
date = []
cit = []

print("Beginning Data Retrieval")
print("-----------------------------")

# Counter to print city numbers
cityNum = 1

# For loop to interate through the cities and request the data from the API
for city in cities:
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&units=imperial&appid={weather_api_key}"
    
    # Make sure there is data to collect on the city
    try:
        # Get the reponse from the API
        response = requests.get(url).json()
        # Load variables with the necessary data
        lat.append(response["coord"]["lat"])
        lon.append(response["coord"]["lon"])
        maxTemp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        cloudiness.append(response["clouds"]["all"])
        windSpeed.append(response["wind"]["speed"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        cit.append(city)
        # Print the record of the city being proccessed
        print(f"Processing Record {cityNum} | {city}")
        # Update the counter for the next city
        cityNum += 1
    # Output line if the city is not found
    except:
        print("City not found. Skipping...")
        
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create a data frame with the collected data on the cities
citiesData = pd.DataFrame({"City": cit, "Lat": lat, "Lng": lon, "Max Temp": maxTemp, "Humidity": humidity,
                         "Cloudiness": cloudiness, "Wind Speed": windSpeed, "Country": country, "Date": date})
# Display the data frame
citiesData.head()
```

```python
# Display summary statisitcs
citiesData.describe()
```

```python
# Export the city data into a .csv
citiesData.to_csv(output_data_file, index=False)
```

## Linear Regression

```python
# Create data frames with the northern and southern hemisphere data
northData = citiesData.loc[citiesData["Lat"]>=0]
southData = citiesData.loc[citiesData["Lat"]<0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Make scatter plot
northData.plot(kind="scatter", x="Lat", y="Max Temp", 
               title="Northern Hemisphere City Latitude vs. Max Temperature (3/16/21)", s=35)
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

# Calcuate lineaer regression equation
(slope, intercept, rvalue, pvalue, sterr) = linregress(northData["Lat"], northData["Max Temp"])
line_eq = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))

# Add line and equation to plot
regressValues = (slope*northData["Lat"]) + intercept
plt.plot(northData["Lat"], regressValues, "red")
plt.annotate(line_eq, (0, -40), fontsize=15, color="red")

# Save scatter plot
plt.savefig("output_data/north_lat_vs_temp.png")
print("The r-value is:", rvalue)
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Make scatter plot
southData.plot(kind="scatter", x="Lat", y="Max Temp", 
               title="Southern Hemisphere City Latitude vs. Max Temperature (3/16/21)", s=35)
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

# Calcuate lineaer regression equation
(slope, intercept, rvalue, pvalue, sterr) = linregress(southData["Lat"], southData["Max Temp"])
line_eq = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))

# Add line and equation to plot
regressValues = (slope*southData["Lat"]) + intercept
plt.plot(southData["Lat"], regressValues, "red")
plt.annotate(line_eq, (-30, 50), fontsize=15, color="red")

# Save scatter plot
plt.savefig("output_data/south_lat_vs_temp.png")
print("The r-value is:", rvalue)
plt.show()
```

The linear regression in this pair of plots is modeling the relationship between a cities latitude and its maximum temperature on a given day in its respective hemisphere. The trend lines show that there is a negative correlation between these variables for the norther hemisphere and a positive correlation in the southern hemisphere. This indicates that the father north, or south a city is located from the equator, the maximum temperature will decrease.
